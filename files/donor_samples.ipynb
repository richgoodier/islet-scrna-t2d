```python
donor_dict = {}
for ID in donor_IDs:
    donor_dict[ID] = 0
```

```python
for sample_title in sample_titles:
    donor_dict[sample_title.split("_")[0]] += 1
```

```python
donor_dict
```

### Randomly Sample 10 from Each Donor

```python
chosen_samples = df.iloc[0:0]
chosen_samples
```

```python
for donor_ID in donor_IDs:
    mask = df["sample_title"].str.contains(donor_ID)
    donor_samples = df[mask]
    random_samples = donor_samples.sample(n=10, random_state=42)
    chosen_samples = pd.concat([chosen_samples, random_samples], ignore_index=True)
```

### Create shell Script

```python
with open("download_all_fastq.sh", "w", encoding='utf-8') as f:
    for link in links:
        f.write(f"wget -nc ftp://{link}\n")
```
