```python
import numpy as np
import matplotlib.pyplot as plt
import csv
import os
```

```python
with open("donor_IDs.csv", 'r', encoding='utf-8') as f:
    donors = f.readlines()

for i in range(len(donors)):
    donors[i] = donors[i].strip()

donors = donors[1:]

donors
```

```python
with open("read_counts.csv", "r", newline="", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    read_counts = [row for row in reader]

read_counts[:5]
```

```python
donor_counts = {}
for donor in donors:
    donor_counts[donor] = []
```

```python
for file in read_counts:
    filename = file["filename"]
    for donor in donors:
        if donor in filename:
            donor_counts[donor].append(file["read_counts"])

    
```
